# How to run: 
1. Specify variables J,z,t,s,t_individual, and their lenghts
2. Using that the lenght of the tensorproduct of J and z, put that number of zeros in Y
3. Execute

```python
z1 = np.append(z, np.random.random_sample((1000,)))
J1 = np.append(J, np.arange(100,300,2.0))

#standard scaling
# J1 = J.reshape((len(J1), 1))
# scaler = MinMaxScaler()
# scaler.fit(J1)
# J1 = scaler.transform(J1)
# print(J1.shape)

X = []
X = [[j, Z] for Z in z for j in J]
#X = [[j, Z] for Z in z1 for j in J1]

X = np.array(X)
print(X)

Y = np.zeros((X.shape[0],1))
print(Y.shape)
print(np.kron(J,z).shape[0])
```

```python
import tensorflow as tf

from tensorflow.keras import datasets, layers, models, Input
import matplotlib.pyplot as plt

model = models.Sequential()
model.add(Input(shape=(2)))
#study activation functions
model.add(layers.Dense(5, activation='relu'))
model.add(layers.Dense(3, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid')) #make sure its between 0 and 1
```

```python
# loss_fn = Custom_Loss_Function()
lr_schedule = keras.optimizers.schedules.ExponentialDecay(
    initial_learning_rate=0.1,
    decay_steps=10000,
    decay_rate=0.9)
opt = keras.optimizers.Adagrad(learning_rate=lr_schedule)
model.compile(optimizer =opt, loss=Custom_Loss_Function, metrics=['accuracy'])
```

```python
history_callback = model.fit(X, Y, epochs=15000, batch_size=np.kron(J,z).shape[0], verbose=2)
loss_history = history_callback.history["loss"]
```

```python
#model.save("my_model")
loss_history = np.array(loss_history)
A = loss_history[-1]
print(A)
A = np.reshape(1,1)
np.savetxt("A10^4_B10^6/loss.txt", A,fmt='%1.4e')
```

```python
#test cases
model.predict(X)
a_file = open("A10^4_B10^6/F.txt", "w")
for row in model.predict(X):
    np.savetxt(a_file, row)
a_file.close()
```
